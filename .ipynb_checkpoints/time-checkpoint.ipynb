```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

# Data preparation

```python
csl_data = pd.read_csv("data/csl-data.csv", na_values=['-1'])

csl_data = csl_data.drop(['eCO2', 'tVOC'], axis=1)
csl_data = csl_data.dropna() #remove rows without AQI data
csl_data = csl_data.drop('timestamp', axis=1) #this is GMT, we want BST
csl_data['Date'] = (csl_data['month'].astype(str)+'/'+csl_data['day'].astype(str)+'/'+
                   csl_data['year'].astype(str)+' '+csl_data['hour'].astype(str)+':'+
                   csl_data['minute'].astype(str))
csl_data['Date'] = pd.to_datetime(csl_data['Date'])
csl_data.head() #get date from BST columns
```

# Overall change

```python
start, end = '2020-06-27 20:00:00', '2020-05-07 06:00:00'

long_data = csl_data.set_index('Date').resample('12H').mean().reset_index()
long_data = long_data.melt(id_vars=['Date'], value_vars=['AQI-1hr-index', 'AQI-std-index'])

sns.set_style("whitegrid")
sns.set_context("notebook")

plt.figure(figsize=(10, 6))
plt.title('12 hour averages')
sns.lineplot(data=long_data, x='Date', y='value', hue='variable');
```

# Time of day

```python
long_data = csl_data.melt(id_vars=['hour'], value_vars=['AQI-1hr-index', 'AQI-std-index'])

plt.figure(figsize=(10, 6))
plt.title("Average AQI over hours of day")
plt.ylim((0, 32))
plt.xlim(0, 23)
sns.lineplot(data=long_data, x='hour', y='value', hue='variable');
```

The one hour index varies more over the day than the standard index because it uses shorter averaging periods so can change more quickly.

We can see that there is a small peak at 9 AM from rush hour and then a decrease between 9 AM and 4 PM. From around 5 PM onwards the index increases as people travel home, peaking at 1 AM when the pollution begins to dissipate and decrease.

The peaks might start to match more closely with rush hour times as increasing numbers of people travel to and from work/school as lockdown eases.

# Day of week

```python
csl_data['weekday'] = csl_data['Date'].dt.dayofweek
long_data = csl_data.melt(id_vars=['weekday'], value_vars=['AQI-1hr-index', 'AQI-std-index'])
plt.figure(figsize=(10, 6))
plt.title("Average AQI over days of week")
plt.xlim(0, 6)
ax = sns.barplot(data=long_data, x='weekday', y='value', hue='variable', alpha=0.7)
ax.set(xticklabels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.xlabel("Day");
```

We can see that the average air quality varies a little over different days of the week. It is highest on Mondays and Tuesdays and lowest on Saturdays.

The difference between weekdays and the weekend may grow as more people travel to work/school as restrictions ease.

# Month

```python
long_data = csl_data.melt(id_vars=['month'], value_vars=['AQI-1hr-index', 'AQI-std-index'])
june_index=csl_data[csl_data.month==6]['AQI-1hr-index']
july_index=csl_data[csl_data.month==7]['AQI-1hr-index']
august_index=csl_data[csl_data.month==8]['AQI-1hr-index']



plt.figure(figsize=(10, 6))
plt.xlim((0, 120))
plt.ylabel("Frequency density")
plt.xlabel("One hour air quality index")
sns.distplot(a=june_index, color="blue", label="June")
sns.distplot(a=july_index, color="green", label="July")
sns.distplot(a=august_index, color="red", label="August", bins=20)
plt.legend()

plt.figure(figsize=(8, 6))
plt.title("Average AQI over different months")
sns.barplot(data=long_data, x='month', y='value', hue='variable', alpha=0.7);
```

We can see that the air quality index decreased slightly between June and July before increasing hugely in August.

This increase may have been partially fuelled by the removal of the law requiring people to work from home and the reopening of cinemas, beauty salons, tattooists, nail bars, spas, tanning shops, museums and art galleries on the 27th of July as well as the reopening of gyms, swimming pools and soft play areas on the 10th of August.
